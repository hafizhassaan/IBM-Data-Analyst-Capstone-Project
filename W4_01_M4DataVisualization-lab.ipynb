## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

table_name = "DataBaseDesireNextYear"  # the table you wish to describe

QUERY = """
SELECT DatabaseDesireNextYear, count(*) as counts
FROM {}
GROUP BY DatabaseDesireNextYear
ORDER BY counts DESC
LIMIT 5
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)

# Create the chart
labels = df['DatabaseDesireNextYear']
sizes = df['counts']
explode = (0.05, 0.01, 0.01, 0.01, 0.01)
plt.pie(
    sizes, explode=explode, labels=labels, autopct="%1.1f%%"
)

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

table_name = "master"  # the table you wish to describe

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM '{}'
WHERE Age BETWEEN 30 AND 35
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)

medians = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

ax = medians.plot(kind='bar', stacked=True)
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='upper left')
ax.tick_params(axis='x', rotation=0)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

table_name = "master"  # the table you wish to describe

QUERY = """
SELECT ConvertedComp, Age FROM '{}'
WHERE Age BETWEEN 45 AND 60
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)

median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Line Chart of Median ConvertedComp for Ages 45-60')
plt.grid(True)
plt.show()
```

```python
# for quiz

table_name = "master"  # the table you wish to describe

QUERY = """
SELECT ConvertedComp, Age FROM '{}'
WHERE Age BETWEEN 25 AND 30
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)

median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Line Chart of Median ConvertedComp for Ages 45-60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

table_name = "master"  # the table you wish to describe

QUERY = """
SELECT MainBranch FROM '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)

main_branch_counts = df['MainBranch'].value_counts()

main_branch_mapping = {
    'I am a developer by profession': 'Professional Developer',
    'I am not primarily a developer, but I write code sometimes as part of my work': 'Non-primary Coder'
}

main_branch_counts.index = main_branch_counts.index.map(main_branch_mapping)

sns.barplot(x=main_branch_counts.values, y=main_branch_counts.index, orient='h')

plt.xlabel('Number of Respondents')
plt.ylabel('MainBranch')
plt.title('Horizontal Bar Chart of MainBranch')
plt.show()
```
